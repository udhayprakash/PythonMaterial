```python
# import library and load data
import pandas as pd
```

```python
# load data from './data/imdb_movie_cleanup.csv'
df_imdb = pd.read_csv('./data/imdb_movie_cleanup.csv')
df_imdb.head()
```

# Rename columns

```python
df_imdb.columns = df_imdb.columns.str.lower().str.replace(' ','_')
df_imdb.head(3)
```

# Drop columns

```python
drop_columns = ['director_name']
df_imdb.drop(drop_columns, axis=1, inplace=True)
df_imdb.head()
```

# Handle missing data

```python
df_imdb.isnull()
```

```python
df_imdb.isnull().sum()
```

```python
# notnull()
df_imdb = df_imdb[df_imdb['duration'].notnull()]
df_imdb.isnull().sum()
```

```python
# fillna
df_imdb.fillna(df_imdb.mean(), inplace=True)
df_imdb.isnull().sum()
```

# Handle duplicate data

```python
df_imdb.duplicated()
```

```python
df_imdb.duplicated().sum()
```

```python
df_imdb.drop_duplicates(keep='first', inplace=True)
df_imdb.duplicated().sum()
```

# Modify column data

```python
import math
def cleanup_title_year(ty):
    if math.isnan(ty):
        return math.nan
    else:
        return int(ty)
```

```python
df_imdb.title_year = df_imdb.title_year.apply(cleanup_title_year)
df_imdb.title_year.head(3)
```
