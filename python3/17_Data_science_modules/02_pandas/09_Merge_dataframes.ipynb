#### Loading libraries

```python
import pandas as pd
```

#### Load customer dataset

```python
df_customers = pd.read_csv('../../datasets/customers.csv')
df_customers
```

#### Load and show orders dataset

```python
df_orders = pd.read_csv('../../datasets/orders.csv')
df_orders
```

### Merge two tables base on same column 'customer_id'
    - Ref: http://www.sql-join.com/

```python
df_merged = pd.merge(left=df_customers,
                     right=df_orders, 
                     on='customer_id',
                     how='inner')
df_merged
```

##### so which customer have largest amount spent ?
##### use groupby then sum on 'amount' column

```python
df_merged.groupby('customer_id').sum()['amount']
```
