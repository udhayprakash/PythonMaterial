## Introduction 
        Files with binary data, bytes that represent non-text information like images, 
        can be easily corrputed when being transferred and processed to text-only systems.
        
        
        Base64 encoding allows us to convert bytes containing binary or text data to ASCII characters. 
        By encoding our data, we improve the chances of it being processed correctly by various systems.


        

## Base64 Encoding 
    - Type of conversion of bytes into ASCII characters. 
    - In mathematics, the base of a number system refers to how 
      manay different characters represent nuhmbers
    - The Base64 character set contains: 
        - 26 uppercase letters
        - 26 lowercase letters 
        - 10 numbers 
        - + and / for new lines (some implementations may use different characters)
        
    - When computer converts base64 characters to binary, each base64 character represents 6 bits of information. 
    
    
__NOTE:__ This is not encryption algorithm. 

```python
for each_chr in "PYTHON":
    print(each_chr, ord(each_chr))
```

```python
import base64

for each_chr in "PYTHON":
    print(each_chr, base64.b64encode(each_chr.encode("utf-8")))
```

```python
for i in range(128):
    print(f'{i:2}, {chr(i)}, {base64.b64encode(chr(i).encode("utf-8"))}')
```

Encodes byte-Strings only 

```python
encoded = base64.b64encode(b"data to be encoded")
print(f"encoded     : {encoded}")

original_msg = base64.b64decode(encoded)
print(f"original_msg: {original_msg}")
```

```python
encoded = base64.b64encode("data to be encoded".encode("ascii"))
print(f"encoded     : {encoded}")

original_msg = base64.b64decode(encoded)
print(f'original_msg: {original_msg.decode("ascii")}')
```

```python
encoded = base64.b64encode("data to be encoded".encode("utf-8"))
print(f"encoded     : {encoded}")

original_msg = base64.b64decode(encoded)
print(f'original_msg: {original_msg.decode("utf-8")}')
```

### URL and Filename Safe Base64 Encoding


```python
data = "abc123!?$*&()'-=@~"

urlSafeEncodedBytes = base64.urlsafe_b64encode(data.encode("utf-8"))
urlSafeEncodedStr = str(urlSafeEncodedBytes, "utf-8")

print(f"{data                =}")
print(f"{urlSafeEncodedBytes =}")
print(f"{urlSafeEncodedStr   =}")
```

```python
decodedBytes = base64.urlsafe_b64decode(urlSafeEncodedBytes)
decodedStr = str(decodedBytes, "utf-8")


print(f"{data         =}")
print(f"{decodedBytes =}")
print(f"{decodedStr   =}")
```

```python
data = "hello world123!?$"

encodedByte = base64.urlsafe_b64encode(data.encode("utf-8"))
encodedStr = str(encodedByte, "utf-8")  # "aGVsbG8gd29ybGQxMjMhPyQ="

decodedBytes = base64.urlsafe_b64decode(encodedStr)
decodedStr = str(decodedBytes, "utf-8")

print(f"orginal message= {data}")
print(f"encoded string = {encodedStr}")
print(f"decoded message= {decodedStr}")  # hello world123!?$
```
